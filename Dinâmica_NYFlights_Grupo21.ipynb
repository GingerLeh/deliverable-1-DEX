```python
#Etapa 1

df_filtro = df[df["dep_delay"]>120]
df_filtro.head()
```

```python
df_describe = df_filtro.describe().T.copy()
df_describe
```

```python
df_describe["amplitude"] = df_describe['max'] - df_describe['min']
df_describe["vari√¢ncia"] = df_describe['std'] **2
df_describe
```

```python
#Etapa 3

lista_destino = ['ORD', 'FLL', 'MCO', 'BOS', 'ATL']
df_med_piores = df_atraso[df_atraso.dest.isin(lista_destino)]
df_med_piores
```

```python
df_med_piores['velocidade m√©dia'] = df_med_piores["distance"]/df_med_piores["air_time"]
df_med_piores
```
